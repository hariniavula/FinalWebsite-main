```python
import pandas as pd

# Load the dataset
df = pd.read_csv("shopping_behavior_updated.csv")

# 1. Check for missing/null values
print(df.isnull().sum())
print("\n")

# 2. Verify numerical columns are in appropriate numerical formats by checking data types
df.info()
print("\n")

# 3. Check consistency for categorical variables (Location, Category, Item Purchased)
categorical_cols = ['Location', 'Category', 'Item Purchased']
for col in categorical_cols:
    print(f"Column: {col}, Number of Unique Values: {df[col].nunique()}")
print("\n")

# 4. Make sure there's no duplicates
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```
